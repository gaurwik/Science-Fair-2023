```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dataFrame = pd.DataFrame(pd.read_csv("/Users/gaurwik/Documents/Science_Fair_2023/calenviroscreen-3.0-results-june-2018-update.csv"))
numeric_only = dataFrame._get_numeric_data()
numeric_only
```

```python
def isExcluded(row):
    excludedWords = {"Score", "Pctl", "CES 3.0", "Pop. Char.", "Longitude", "Latitude", "Housing Burden", "Pollution Burden"}
    for i in excludedWords:
        if (i in row[0][0] or i in row[0][1]):
            return False
    
    return row[0][0] != row[0][1]
```

```python
corr = dataFrame.corr()
c1 = corr.abs().unstack()
c1 = c1.drop_duplicates()
c1 = c1.sort_values(ascending=False)

oldDf = pd.DataFrame(c1)
df = pd.Series
first = True
for row in oldDf.iterrows():
    if (isExcluded(row)):
        x = pd.Series([row[0][0], row[0][1], row[1]])
        if (first):
            df = x
            first = False
        else:
            df = pd.concat([df, x], axis = 1, ignore_index=True)

df = pd.DataFrame(df)
df = df.transpose()
display(df)
```

```python
health = {"Asthma", "Low Birth Weight", "Cardiovascular Disease"}
healthDf = pd.Series
first = True
for row in df.iterrows():
    if (row[1][0] in health or row[1][1] in health):
        x = pd.Series([row[0], row[1][0], row[1][1], row[1][2]])
        if (first):
            healthDf = x
            first = False
        else:
            healthDf = pd.concat([healthDf, x], axis = 1, ignore_index = True)

healthDf = pd.DataFrame(healthDf)
healthDf = healthDf.transpose()
display(healthDf)
```
